```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")
df.head(5)
```

### TASK 1: Plot the FlightNumber vs. PayloadMass and overlay the outcome of the launch


```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect=5)
plt.xlabel("Flight Number",fontsize=14)
plt.ylabel("Pay load Mass (kg)",fontsize=14)
plt.show()
```

As the flight number increases, the first stage is more likely to land successfully.

### TASK 2: Visualize the relationship between Flight Number and Launch Site


```python
sns.catplot(y="LaunchSite", x="FlightNumber", hue="Class", data=df, aspect=1.5)
plt.ylabel("Launch Site", fontsize=14)
plt.xlabel("Flight Number", fontsize=14)
plt.show()
```

Different launch sites have different success rates.  <code>CCAFS LC-40</code> has a success rate of 60 %, while  <code>KSC LC-39A</code> and <code>VAFB SLC 4E</code> has a success rate of 77%.


### TASK 3: Visualize the relationship between Payload and Launch Site


```python
sns.catplot(y="LaunchSite", x="PayloadMass", hue="Class", data=df, aspect=1.5)
plt.ylabel("Launch Site", fontsize=14)
plt.xlabel("Payload Mass", fontsize=14)
plt.show()
```

For the VAFB-SLC launchsite there are no rockets launched for heavypayload mass (greater than 10000).

### TASK  4: Visualize the relationship between FlightNumber and Orbit type


```python
sns.catplot(y="Orbit", x="FlightNumber", hue="Class", data=df, aspect=1)
plt.ylabel("Orbit",fontsize=15)
plt.xlabel("FlightNumber",fontsize=15)
plt.show()
```

In the LEO orbit the Success appears related to the number of flights; on the other hand, there seems to be no relationship between flight number in GTO orbit.

### TASK  5: Visualize the relationship between Payload and Orbit type


```python
sns.catplot(y="Orbit", x="PayloadMass", hue="Class", data=df, aspect = 1)
plt.ylabel("Orbit",fontsize=15)
plt.xlabel("Payload Mass",fontsize=15)
plt.show()
```

With heavy payloads the positive landing rate are more for Polar, VLEO and ISS. However for GTO it is hard to distinguish this well as both positive and negative landing rates are there.

### TASK  6: Visualize the launch success yearly trend


```python
date = pd.to_datetime(df['Date'])
year = date.dt.year
```

```python
sns.lineplot(x = year, y = 'Class', data = df)
```

The sucess rate since 2013 kept increasing till 2020
