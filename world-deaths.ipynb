2. Which 2 continents have the highest number of deaths caused by child wasting and how this is changing over 5-year periods?

```python
sql = '''
SELECT continent, SUM(`Child wasting`)/1000000 AS `Deaths (mln)`
FROM deaths d
JOIN continents c
	ON d.entity = c.country
WHERE Year BETWEEN {s} AND {e}
GROUP BY continent
ORDER BY continent ASC
LIMIT 2;
'''

sy0 = 1990 # starting year
ey0 = 2017 # ending year
every = 5
interval = int((ey0 - sy0)/every)

labels = []
lab = []
cw_con1 = []
cw_con2 = []
for i in range(1,interval+1):
    sy = sy0+(i-1)*every
    ey = sy0+i*every
    labels.append(str(sy)+'-'+str(ey))

    c.execute(sql.format(s=sy, e=ey))
    ans = c.fetchall()
    print(ans)
    if i == 1:
        lab.append(ans[0][0])
        lab.append(ans[1][0])
    cw_con1.append(ans[0][1])
    cw_con2.append(ans[1][1])
```

Which top 5 countries had the highest deaths per mln people reduce in top 1 continent? Show the difference between the highest and the lowest deaths per mln people.

```python
sql = '''
SELECT p.country, MAX(`Child wasting`/value*1000000)-MIN(`Child wasting`/value*1000000) AS `Death rate change`
FROM population p
JOIN deaths d
	ON d.entity = p. country AND d.year = p. year
JOIN continents c
	USING (country)
WHERE continent = '{}'
GROUP BY country
ORDER BY `Death rate change` DESC
LIMIT 5;
'''

country = []
diff = []
c.execute(sql.format(lab[1]))
ans = c.fetchall()

for el in ans:
	country.append(el[0])
	diff.append(el[1])
```

```python
x = np.arange(1,interval+1)
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,5))

w = 0.25
A = ax1.bar(x-w/2, cw_con1, width=w, color='brown', label=lab[0])
B = ax1.bar(x+w/2, cw_con2, width=w, color='orange', label=lab[1])
ax1.set_xlabel('Interval', fontsize=14)
ax1.set_ylabel('Deaths (mln)', fontsize=14)
ax1.set_title('Deahts in terms of child wasting', style='italic')
ax1.set_xticks(x)
ax1.set_xticklabels(labels)
ax1.legend()
ax1.bar_label(A, fmt='%.1f')
ax1.bar_label(B, fmt='%.1f')

C = ax2.bar(country, diff, width=w*2, color='orange')
ax2.set_xlabel('Country', fontsize=12)
ax2.set_ylabel('Deaths', fontsize=12)
ax2.set_title('Difference in deahts caused by\nchild wasting per mln inhabitants in '+lab[1], style='italic')
ax2.bar_label(C, fmt='%.0f')

plt.show()
```

3. What is the trend in Baltic countries deaths in terms of alcohol use and drug use after 2000?

```python
sql = '''
SELECT year, SUM(`Alcohol use`), SUM(`Drug use`)
FROM deaths d
JOIN continents c
	ON c.country = d.entity
WHERE country = '{}' AND year >= 2000
GROUP BY country, year
ORDER BY country ASC, year ASC;
'''
baltics = ['Lithuania', 'Latvia', 'Estonia']
colors = ['black', 'green', 'red']

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(8,5))

for el in baltics:
    c.execute(sql.format(el))
    ans = c.fetchall()

    year = []
    alco = []
    drug = []
    for eil in ans:
        year.append(eil[0])
        alco.append(eil[1])
        drug.append(eil[2])

    ax1.plot(year, alco, lw=2, marker='o', ms=5, color=colors[baltics.index(el)])
    ax2.plot(year, drug, lw=2, marker='x', ms=8, color=colors[baltics.index(el)])

ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('Deahts', fontsize=12)
ax1.set_title('Deahts in Baltic countries\nin terms of alcohol use', style='italic')
ax1.legend(baltics)

ax2.set_xlabel('Year', fontsize=12)
ax2.set_ylabel('Deahts', fontsize=12)
ax2.set_title('Deahts in Baltic countries\nin terms of drug use', style='italic')
ax2.legend(baltics)
ax2.set_facecolor('#ffffe4')

plt.tight_layout()
plt.show()
```

4. How the continents are divided in terms of deaths from smoking (average death rates in period 1990-2017)?

```python
sql = '''
SELECT continent, AVG(death_rate) AS `death rate` FROM
	(SELECT continent, d.year, SUM(smoking)/SUM(value) as death_rate
	FROM deaths d
	JOIN population p
		ON d.entity = p.country AND d.year = p.year
	JOIN continents c
		ON c.country = d.entity
	GROUP BY continent, d.year) d
GROUP BY continent
ORDER BY `death rate`;
'''
smo = pd.read_sql_query(sql, con=DB)
smo
```

```python
colors = ['#febc68', '#fcdb8c', '#1f9589', '#41c0b5', '#434247', '#5c5a66']

fig, ax = plt.subplots()
ax.pie(
    smo['death rate'],
    labels = smo['continent'],
    autopct = '%.1f%%',
    startangle = 90,
    explode = [0,0,0,0,0,0.1],
    wedgeprops = {'linewidth': 2.0, 'edgecolor': 'white'},
    colors = colors
    )
ax.set_title('Average death rate in terms of smoking\nby continents in period 1990-2017', style='italic')
```

5. Which countries are in the top 5 in terms of deaths from smoking per million inhabitants in each continent (average death rates in period 1990-2017)?

```python
sql = '''
SELECT continent, country, AVG(death_rate)*1000000 AS `deaths` FROM
	(SELECT continent, c.country AS country, smoking/value AS death_rate
	FROM deaths d
	JOIN population p
		ON d.entity = p.country AND d.year = p.year
	JOIN continents c
		ON c.country = d.entity) d
GROUP BY continent, country
HAVING continent = '{}'
ORDER BY `deaths` DESC
LIMIT 5;
'''
```

```python
fig, ax = plt.subplots(2,3, figsize=(16,10))

xi = 0
yi = 0
for con in range(0,len(smo)):
    smo_cou = pd.read_sql_query(sql.format(smo.loc[con][0]), con=DB)
    A = ax[yi][xi].bar(smo_cou['country'], smo_cou['deaths'], color=colors[con], width=0.4)
    ax[yi][xi].set_xticks(smo_cou['country'])
    ax[yi][xi].set_xticklabels(smo_cou['country'], rotation=45)
    ax[yi][xi].set_title('5 countries by deaths in terms of smoking per mln\ninhabitants in '+smo.loc[con][0]+' in period 1990-2017', style='italic')
    ax[yi][xi].bar_label(A, fmt='%.0f')
    ax[yi][xi].set_xlabel('Country', fontsize=12)
    ax[yi][xi].set_ylabel('Deaths', fontsize=12)
    ax[yi][xi].set_ylim(0, smo_cou['deaths'].max()*1.1)
    xi = xi+1
    if xi == 3:
        xi = 0
        yi = yi+1

plt.tight_layout()
plt.show()
```

6. Which Asian countries (top 5) have the most deaths per million inhabitants in term of air pollution in 2017

```python
sql = '''
SELECT c.country, `Air pollution`/value*1000000 AS `death rate`
FROM deaths d
JOIN continents c
	ON d.entity = c.country
JOIN population p
	ON d.entity = p.country AND d.year = p.year
WHERE continent = 'Asia' AND d.year = 2017
ORDER BY `death rate` DESC
LIMIT 5;
'''
air = pd.read_sql_query(sql, con=DB)

fig, ax1 = plt.subplots()
air.plot(kind='barh', x='country', y='death rate', ax=ax1, color='xkcd:dusky purple', legend=None)
ax1.set_xlabel('Deaths', fontsize=12)
ax1.set_ylabel('Country', fontsize=12)
ax1.invert_yaxis()
plt.title('Deaths per million inhabitants in terms of air pollution in Asia', style='italic')
plt.show()
```

### Insights:
1. Even though the World population was increasing during the given period, deaths by unsafe water source were decreasing. This could be due to improved technologies for water cleaning devices, improved medicine or changed strategy for death registration (more risk factors introduced).

2. Asia and Africa had the biggest number of deaths by child wasting, but the numbers were decreasing over the years, especially is Asia, which now even has lower deaths number than Africa. The biggest decrease in number of deaths per million inhabitants in Asia made North Korea. Improved agriculture, political decisions, war against hunger were the main effects for the decrease in number of deaths for both continents.

4. Deaths by alcohol use and drug use were increasing in Baltic countries from 2000 to 2007, especially in Lithuania. The reason for that could be the increasing economy, but poor education about psychotropic substances. After 2008 there was a big drop in deaths for Lithuania and slower drop in deaths for Latia and Estonia. The reason for that could the global financial crisis and housing bubble, which led to shrinking economy and not being able to buy everything you want.

5. The continent with the highest death rate in terms of smoking is Europe and the country with the biggest number of deaths per million inhabitants is Serbia. In Asia the country with the biggest number is Russian Federation. The continent with the lowest death rate is Africa.

6. North Korea has the biggest number of deaths per million inhabitants in terms of air pollution because of authorities failing to combat fine dust. India and China are the industrial leaders and are known as the most air polluted countries. In India most of the air pollution comes from dust, construction and waste burning. China even though invest in green energy sources continues to fulfill its growing demand for energy with fossil fuels.
